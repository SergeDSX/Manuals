## Объект plot (линейный график)

Используется для отрисовки **линейных графиков**. Если в него не передавать значения, то будет отрисовано пустое поле. Если передать вектор, то его значения будут значениями оси Y, а на оси Х - индексы этих значений. 

```python
values = {'x': [2, 4, 8, 12, 15, 20, 25], 'y': [0, 5, 4, 11, 12, 10, 9]}
df = pd.DataFrame(values)

fig, axes = plt.subplots(1, 3, figsize = (16, 3))
axes[0].plot()
axes[1].plot(df['x'])
axes[2].plot(df['x'], df['y'])
plt.show()
```

## Объект grid (сетка графика)

Добавление параметра **grid()** позволяет отобразить сетку графика. По умолчанию, как видно выше, она отображаться не будет. 

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].plot()
axes[1].grid()
axes[1].plot()
plt.show()
```

### Параметры объекта grid

* **color** - цвет сетки. 
* **linewidth** - толщина. 
* **linestyle** - стиль линий. [ '-' | '--' | '-.' | ':' | 'steps' | ...]. Более подробно о стилях: [тут](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-py). 
* **alpha** - прозрачность. Значение от 0 до 1.
* **axis** - указание, для каких осей применяется сетка. Возможные значения: 'both', 'x', 'y'.

Работа объекта схожа при вызове **plt.grid** и **axes.grid**. В первом случае происходит управление сеткой для последнего созданного графика, во второй для конкретной указанной оси. 

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].grid(color = 'blue')
axes[0].plot()
axes[1].grid(color = 'red')
axes[1].plot()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].grid(linewidth = 0.5)
axes[0].plot()
axes[1].grid(linewidth = 3)
axes[1].plot()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].grid(linestyle = ':', linewidth = 1.5)
axes[0].plot()
axes[1].grid(linestyle = '-.', linewidth = 1.5)
axes[1].plot()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].grid(linestyle = ':', linewidth = 1.5, color = 'green', alpha = 0.3)
axes[0].plot()
axes[1].grid(linestyle = ':', linewidth = 1.5, color = 'green', alpha = 0.9)
axes[1].plot()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].grid(linestyle = ':', linewidth = 1.5, color = 'green', axis = 'x')
axes[0].plot()
axes[1].grid(linestyle = ':', linewidth = 1.5, color = 'green', axis = 'y')
axes[1].plot()
plt.show()
```

## Объект title (заголовок графика)

Если используется для одиночного графика, то настраивается через метод **title**. Если же необходимо настроить заголовок для каждой из осей в случае холста с нескольими графиками, то это делается с использованием метода **set_title**.

Имеет следующие настройки:
* **label** - текст заголовка.
* **color** - цвет.
* **fontsize** - размер шрифта.
* **alpha** - прозрачность текста. Значение от 0 до 1.
* **loc** - расположение. Возможные значения: 'center' (по умолчанию), 'left', 'right'.
* **y** - положение заголовка по оси Y. По умолчанию - 1.
* **pad** - отступ от графика в пикселях.
* **rotation** - угол поворота заголовка. При использовании угла центр заголовка смещается так, чтобы низ подписи был в положении возле графика.

```python
plt.figure(figsize = (16,1))
plt.plot()
plt.title(label = 'Заголовок')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].set_title(label = 'Заголовок', color = 'green')
axes[0].plot()
axes[1].set_title(label = 'Заголовок', color = 'red')
axes[1].plot()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].set_title(label = 'Заголовок', color = 'green', fontsize = 5)
axes[0].plot()
axes[1].set_title(label = 'Заголовок', color = 'green', fontsize = 20)
axes[1].plot()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].set_title(label = 'Заголовок', color = 'red', fontsize = 12, alpha = 0.3)
axes[0].plot()
axes[1].set_title(label = 'Заголовок', color = 'red', fontsize = 12, alpha = 0.9)
axes[1].plot()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].set_title(label = 'Заголовок', color = 'red', fontsize = 12, alpha = 0.7, loc = 'left')
axes[0].plot()
axes[1].set_title(label = 'Заголовок', color = 'red', fontsize = 12, alpha = 0.7, loc = 'right')
axes[1].plot()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].set_title(label = 'Заголовок', color = 'red', fontsize = 12, alpha = 0.7, y = 1)
axes[0].plot()
axes[1].set_title(label = 'Заголовок', color = 'red', fontsize = 12, alpha = 0.7, y = 1.3)
axes[1].plot()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].set_title(label = 'Заголовок', color = 'red', fontsize = 12, alpha = 0.7, pad = 10)
axes[0].plot()
axes[1].set_title(label = 'Заголовок', color = 'red', fontsize = 12, alpha = 0.7, pad = 30)
axes[1].plot()
plt.show()
```

## Объекты xlabel, ylabel (подписи осей графика)

Объекты используются для настройки подписей осей графика. **xlabel** - подпись оси Х, **ylabel** - подпись оси Y. 
Если необходимо настроить их для каждой из осей в случае холста с нескольими графиками, то это делается с использованием методов **set_xlabel**, **set_ylabel**. 

Настройки этих объектов такие же, как и объекта **title**, кроме отступа. Аргумент для его настройки называется **labelpad**, а не **pad**.
Также аргумент **label** будет называться **xlabel** или **ylabel**, в зависимости от функции. 

```python
plt.figure(figsize = (16,1))
plt.plot()
plt.xlabel(xlabel = 'x')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].set_xlabel(xlabel = 'x', color = 'red', fontsize = 16, alpha = 0.7)
axes[0].plot()
axes[1].set_xlabel(xlabel = 'x', color = 'red', fontsize = 16, alpha = 0.7, rotation = 45)
axes[1].plot()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].set_ylabel(ylabel = 'y', color = 'green', fontsize = 12, alpha = 0.7)
axes[0].plot()
axes[1].set_ylabel(ylabel = 'y', color = 'green', fontsize = 20, alpha = 0.7)
axes[1].plot()
plt.show()
```

## Объекты xlim, ylim (ограничение осей графика)

Функции принимают лишь аргументы левой и правой границы оси: **xlim** - для оси Х, **ylim** - для оси Y соответственно.
Они используются для единичного графика. Если необходимо настроить их для каждой из осей в случае холста с нескольими графиками, то это делается с использованием методов **set_xlim**, **set_ylim**. 

```python
plt.figure(figsize = (16,1))
plt.plot()
plt.xlim(0,1)
plt.ylim(0,1)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 3))
axes[0].set_ylim(0,2)
axes[0].plot()
axes[1].set_xlim(-1,1)
axes[1].plot()
plt.show()
```

Для построенного графика границы можно извлечь в переменную. Для единичного графика - через функцию **xlim()**, а для отдельной оси холста - через функцию **get_xlim()**. 

```python
plt.figure(figsize = (16,1))
plt.plot()
plt.ioff()
lim = plt.xlim()
print(f'Границы: {lim}')
```

## Объекты xticks, yticks (деления осей графика)

Функции используются для настройки делений осей и их подписей. 
Параметры объектов: 
* **ticks** - численные значения, которым соответствуют деления осей.
* **labels** - подписи осей. 
* **minor** - принимает значения True / False. По умолчанию False. При значении True подписи осей лучше оставлять по умолчанию, т.к. иначе будет возникать предупреждение, связанное с FixedLocator.
* **rotation** - угол поворота подписи деления.
* **ha** - выравнивание подписи относительно деления. Принимает значения: 'left', 'right', 'center'ю

```python
plt.figure(figsize = (16,1))
plt.plot()
plt.xticks(ticks=[0,1,2,3], labels=['a','b','c','d'])
plt.show()
```

```python
plt.figure(figsize = (16,1))
plt.plot()
plt.xticks(ticks=[0,1,2,3], labels=['a','b','c','d'], rotation=90)
plt.show()
```

Если необходимо настроить деления для каждой из осей в случае холста с нескольими графиками, то это делается с использованием методов **set_xticks** / **set_yticks** и **set_xticklabels** / **set_yticklabels**.

```python
fig, axes = plt.subplots(1, 3, figsize = (16, 3))
axes[0].set_xticks(ticks=[0,1,2,3])
axes[0].plot()
axes[1].set_xticks(ticks=[0,1,2,3], labels=['a','b','c','d'])
axes[1].plot()
axes[2].set_xticks(ticks=[0,1,2,3], labels=['a','b','c','d'], rotation=90)
axes[2].plot()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize = (16, 3))
axes[0].set_xticks(ticks=[0,1,2,3], labels=[10,9,8,7])
axes[0].plot()

axes[1].set_xticks(ticks=[0,1,2,3], minor=True)
axes[1].plot()

axes[2].set_xticks(ticks=[0,1,2,3], labels=['a','b','c','d'], rotation=90)
axes[2].plot()
plt.show()
```

Вместо указанных выше функций настраивать деления осей можно с помощью метода **tick_params**. 
Он имеет те же настройки, но они могут иметь другие наименования (например, **labelrotation** вместо **rotation**). [Документация](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.tick_params.html)

## Настройка линий объекта plot

Линии объекта **plot** имеют многие из тех же параметров, которые имеет объект **grid**.
* **color** (c) - цвет. 
* **linewidth** (lw) - толщина. 
* **linestyle** (ls) - стиль линий. [ '-' | '--' | '-.' | ':' | 'steps' | ...]. Более подробно о стилях: [тут](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-py). Если присвоить этому параметру значение 'None' (строку!), то линия не будет отображаться (может быть полезно, когда нужно отобразить только точки).
* **alpha** - прозрачность. Значение от 0 до 1.

```python
values = {'x': [2, 4, 8, 12, 15, 20, 25], 'y': [0, 5, 4, 11, 12, 10, 9]}
df = pd.DataFrame(values)

fig, axes = plt.subplots(1, 3, figsize = (16, 3))
axes[0].plot(df['x'], df['y'], color = 'green', lw = 1, ls = ':')
axes[1].plot(df['x'], df['y'], color = 'darkred', lw = 2, ls = '--')
axes[2].plot(df['x'], df['y'], color = 'black', lw = 3, alpha = 0.3)
plt.show()
```

Помимо этого, отдельно можно настроить маркеры (точки, соответствующие значениям на графике). Для этого используются параметры:
* **marker** - вид маркера.
* **markersize** (ms) - размер маркера.
* **markeredgecolor** (mec) - цвет границы маркера.
* **markeredgewidth** (mew) - толщина маркера.
* **markerfacecolor** (mfc) - цвет внутренней части маркера.

```python
values = {'x': [2, 4, 8, 12, 15, 20, 25], 'y': [0, 5, 4, 11, 12, 10, 9]}
df = pd.DataFrame(values)

fig, axes = plt.subplots(1, 3, figsize = (16, 3))
axes[0].plot(df['x'], df['y'], ls = 'None', marker = 'x', ms = 5)
axes[1].plot(df['x'], df['y'], ls = 'None', marker = 's', ms = 10, mfc = 'yellow')
axes[2].plot(df['x'], df['y'], ls = 'None', marker = 'o', ms = 15, mec = 'black')
plt.show()
```

```python
values = {'x': [2, 4, 8, 12, 15, 20, 25], 'y': [0, 5, 4, 11, 12, 10, 9]}
df = pd.DataFrame(values)

fig, axes = plt.subplots(1, 3, figsize = (16, 3))
axes[0].plot(df['x'], df['y'], ls = 'None', marker = 'x', ms = 10, mew = 5)
axes[1].plot(df['x'], df['y'], ls = 'None', marker = 's', ms = 10, mew = 5, mfc = 'black')
axes[2].plot(df['x'], df['y'], ls = 'None', marker = 'o', ms = 10, mew = 15, mec = 'black')
plt.show()
```

## Отрисовка нескольких линий на одном линейном графике

Для отображения нескольких линий на одном графике можно:
* в рамках одного объекта **plot** через запятую передавать пары столбцов (x, y);
* создать несколько объектов, в каждом из которых настраивается своя линия с координатами (x, y).

Первый случай удобнее, когда линий немного и нет необходимости применять специальные настройки к каждой линии. Все настройки будут распространяться на все линии. Во втором случае можно отдельно настроить каждую линию. 

Если не вызывать **plt.show**, то графики будут показаны вместе. Когда их надо показать по отдельности, то после каждого необходимо вызвать этот метод.

```python
values = {'x1': [2, 4, 8, 12, 15, 20, 25], 
          'x2': [1, 3, 7, 10, 13, 18, 22],
          'y1': [0, 5, 4, 11, 12, 10, 9],
          'y2': [3, 10, 15, 6, 5, 2, 1]}

df = pd.DataFrame(values)
plt.figure(figsize = (16,3))
plt.plot(df['x1'], df['y1'], df['x2'], df['y2'], df['x2']+2, df['y2']+2, linewidth=3, marker='o')
plt.show()
```

```python
plt.figure(figsize = (16,3))
plt.plot(df['x1'],   df['y1'],   lw=3, marker='o')
plt.plot(df['x2'],   df['y2'],   lw=4, marker='^')
plt.plot(df['x2']+2, df['y2']+2, lw=5, marker='s')

plt.show()
```

## Объект twin(x/y) (добавление вспомогательных осей)

Вспомогательная ось создаётся только для оси, созданной как **subplot**. 
Для этой оси задаётся график, после чего в новую переменную сохраняется метод этого графика **twinx** (если вспомогательная ось - это ось y).
Далее этой переменной присваивается график, который должен быть создан для вспомогательной оси. 

```python
values = {'x1': [2, 4, 8, 12, 15, 20, 25], 
          'x2': [1, 3, 7, 10, 13, 18, 22],
          'y1': [0, 5, 4, 11, 12, 10, 9],
          'y2': [3, 10, 15, 6, 5, 2, 1]}

df = pd.DataFrame(values)

fig, ax1 = plt.subplots(figsize = (16,3))
ax1.plot(df['x1'],   df['y1'], marker='o', c = 'darkred')
ax2 = ax1.twinx()
ax2 = plt.plot(df['x2'], df['y2'], marker='^')
plt.show()
```

Если необходимо добавить вспомогательную ось Y, то используется метод **twiny**. Далее по тому же принципу.

```python
fig, ax1 = plt.subplots(figsize = (16,3))
ax1.plot(df['x1'], df['y1'], marker='o', c='darkred')
ax2 = ax1.twiny()
ax2 = plt.plot(df['x2'], df['y2'], marker='^')
plt.show()
```

## Объект legend (легенда графика)

Легенда графика показывается только, если вызван метод **legend**. Он имеет следующие аргументы:
* **handles** - параметры линий, отображаемых в легенде.
* **labels** - подписи легенды.
* **loc** - расположение блока легенды.
* **bbox_to_anchor** - координаты расположения блока легенды.
* **ncols** - количество колонок для размещения элементов легенды.
* **fontsize** - размер шрифта подписей легенды.
* **labelcolor** - цвет подписей.
* **mode** - режим легенды. Возможные значения: "expand" / None. Если присвоено значение "expand", то легенда расширяется, чтобы заполнить всю область относительно оси.
* **title** - заголовок легенды.
* **title_fontsize** - цвет заголовка легенды.
* **frameon** - рамка легенды. Возможные значения: True / False. По умолчанию True.

Если необходимо отобразить только подписи легенды, то достаточно в функции **plot** указать аргумент **label**. Вызов **legend** без аргументов вернёт легенду с настройками по умолчанию. Если же **label** не указывать, то можно передать список с наименованиями как аргумент **labels**.

В случае наличия нескольких осей функция **plt.legend** управляет легендой для последней оси, тогда как метод оси **axes.legend** позволяет задавать легенду для каждой оси отдельно.

```python
fig, ax = plt.subplots(1,2, figsize=(16,3))

ax[0].plot(df['x1'], df['y1'], linewidth=3, marker='o', label='x1_y1')
ax[0].plot(df['x2'], df['y2'], linewidth=3, marker='o', label='x2_y2')
ax[1].plot(df['x1'], df['y1'], linewidth=3, marker='o')
ax[1].plot(df['x2'], df['y2'], linewidth=3, marker='o')
ax[0].legend()
ax[1].legend(labels=['x1_y1','x2_y2'])
plt.show()
```

```python
fig, ax = plt.subplots(1,3, figsize=(16,3))

ax[0].plot(df['x1'], df['y1'], df['x2'], df['y2'], linewidth=3, marker='o')
ax[1].plot(df['x1'], df['y1'], df['x2'], df['y2'], linewidth=3, marker='o')
ax[2].plot(df['x1'], df['y1'], df['x2'], df['y2'], linewidth=3, marker='o')
ax[0].legend(labels=['x1_y1','x2_y2'], loc = 'upper left', fontsize = 12, labelcolor = 'green', title = 'Upper Left', title_fontsize = 16)
ax[1].legend(labels=['x1_y1','x2_y2'], loc = 'lower right', fontsize = 16, labelcolor = 'black', title = 'Lower Right', title_fontsize = 14)
ax[2].legend(labels=['x1_y1','x2_y2'], loc = 'center', fontsize = 20, labelcolor = 'darkred', title = 'Center', title_fontsize = 12)
plt.show()
```

Произвольное размещение легенды осуществляется с помощью опции **bbox_to_anchor**. Координата (0,0) - это левый нижний угол, в котором размещается правая верхняя часть блока. Соответственно: 
* если необходимо разместить легенду в левой части, первая координата смещается в отрицательную зону.
* если необходимо разместить её снизу - то вторая координата смещается в отрицательную зону.

```python
fig, ax = plt.subplots(1,3, figsize=(16,3))

ax[0].plot(df['x1'], df['y1'], df['x2'], df['y2'], linewidth=3, marker='o')
ax[1].plot(df['x1'], df['y1'], df['x2'], df['y2'], linewidth=3, marker='o')
ax[2].plot(df['x1'], df['y1'], df['x2'], df['y2'], linewidth=3, marker='o')
ax[0].legend(labels=['x1_y1','x2_y2'], bbox_to_anchor=(0,0), fontsize=12, labelcolor='green', title='1st', title_fontsize=16)
ax[1].legend(labels=['x1_y1','x2_y2'], bbox_to_anchor=(-1,0), fontsize=16, labelcolor='black', title='2nd', title_fontsize=14)
ax[2].legend(labels=['x1_y1','x2_y2'], bbox_to_anchor=(0,-1), fontsize=20, labelcolor='darkred', title='3rd', title_fontsize=12)
plt.show()
```

Когда необходимо разместить легенду выше или правее графика, то необходимо подбирать положительные значения, в зависимости от конфигурации холста.

## Послойное создание графиков с неравномерным размещением на холсте (объект subplot)

С помощью функции **subplot** можно сформировать "слой" холста, и с помощью координат указать размещение конкретного графика на этом слое. 
Координаты (x, y, z), передаются как аргументы функции **subplot**, где x - число строк в слое, y - число столбцов в слое, z - индекс графика. 
На каждом пересечении строки и столбца выделяется место под график, индексация производится построчно. 

После указания каждого места в слое должна вызываться команда, формирующая график, который займёт это место.

```python
# распределение графиков по холсту относительно строк

x = np.linspace(0, 10, 5)
y1, y2, y3, y4 = x**2, x**3, x**0.5, np.sin(x**2)

plt.figure(figsize=(12, 7))

plt.subplot(3, 3, 1)
plt.plot(x, y1, 'r')
plt.subplot(3, 3, 2)
plt.plot(x, y2, 'b')
plt.subplot(3, 3, 3)
plt.plot(x, y3, 'g')

plt.subplot(3, 2, 3)
plt.plot(x, y1, 'm')
plt.subplot(3, 2, 4)
plt.plot(x, y2, 'c')

plt.subplot(3, 1, 3)
plt.plot(x, y4, 'grey')

plt.show()
```

```python
# распределение графиков по холсту относительно столбцов

x = np.linspace(0, 10, 5)
y1, y2, y3, y4 = x**2, x**3, x**0.5, np.sin(x**2)

plt.figure(figsize=(12, 7))

plt.subplot(3, 3, 1)
plt.plot(x, y1, 'r')
plt.subplot(3, 3, 2)
plt.plot(x, y2, 'b')

plt.subplot(2, 3, 4)
plt.plot(x, y1, 'm')
plt.subplot(2, 3, 5)
plt.plot(x, y2, 'c')

plt.subplot(1, 3, 3)
plt.plot(x, y4, 'grey')

plt.show()
```

Атрибуты для каждого графика (подписи, легенда, название) - задаются после соответствующей команды **plot** для этого графика. 

## Объект Figure (настройка холста)

Для холста можно задать следующие настройки:
* **figsize** - размер холста (координаты по горизонтали и вертикали);
* **dpi** - разрешение (точек на дюйм);
* **facecolor** - цвет подложки;
* **edgecolor** - цвет границы;
* **frameon** - рамка холста. Возможные значения True / False.
* **sharex / sharey** - общая ось Х / Y. Возможные значения True / False.

Холст не может быть вызван без осей. Чтобы настроить оси, необходимо использовать функцию **add_axes**.

В неё передаётся список вида - (начальная координата Х, начальная координата Y, ширина, высота). Графики на этих осях можно отобразить либо вызывая для созданной оси метод **plot** (если график вида **plot**), либо присваивая переменной, в которую сохранена ось, этот график.

```python
x = np.linspace(-np.pi, np.pi, 20)
cos_x = np.cos(x) 
sin_x = np.sin(x) 

fig = plt.figure(figsize=(16,3), facecolor='lightgreen')

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([1, 1, 2, 2])
ax2.plot(x, cos_x, color = 'red', lw = 2) 

plt.show()

fig = plt.figure(figsize=(16,3), facecolor='lightgreen')

ax1 = fig.add_axes([0, 3, 1, 1])
ax2 = fig.add_axes([1, 2, 1, 1])
ax2 = plt.plot(x, sin_x, color = 'black', lw = 3) 

plt.show()
```

После создания холста можно задать ему заголовок, и подпись для оси Х / Y с помощью команд **fig.suptitle**, **fig.supxlabel**, **fig.supylabel**. Все эти функции имеют одинаковые аргументы кастомизации:
* **t** - текст подписи.
* **x** - координата размещения по оси Х.
* **y** - координата размещения по оси Y.
* **horizontalalignment / ha** - выравнивание по горизонтали. Возможные варианты: 'center', 'left', 'right', по умолчанию - 'center'.
* **verticalalignment / va**- выравнивание по вертикали. Возможные варианты: 'center', 'left', 'right', по умолчанию - 'center'.
* **fontsize / size** - размер шрифта.

## Объект GridSpec (настройка сетки графиков)

Объект импортируется отдельной командой. 
Для его использования необходимо сначала создать объект **Figure**, а затем связать объект **GridSpec** с объектом **Figure**, как аргумент функции. 
Помимо объекта **Figure**, в аргументах **GridSprec** также указывается число строк и столбцов предполагаемой сетки. 

Далее на созданный холст (**Figure**) добавляются оси с помощью команды **add_suplot** и в аргументе функции указывается область сетки, в которой она будет отображаться (указав координаты строки и столбца).

```python
fig = plt.figure(figsize=(10, 5))
gs = GridSpec(nrows=2, ncols=2, figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

plt.tight_layout()
plt.show()
```

Если указать координаты нескольких строк или столбцов, то произойдёт объединение сетки. 

```python
fig = plt.figure(figsize=(10, 5))
gs = GridSpec(nrows=2, ncols=2, figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1:,0:])

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
gs = GridSpec(nrows=2, ncols=2, figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[0:,1:])

plt.tight_layout()
plt.show()
```

Исходная сетка также может быть преобразована через соотношения между площадями осей. Это настраивается с помощью аргументов **width_ratios** и **height_ratios**. В записи ниже длина 1-го столбца = 1, 2-го столбца = 2, высота 1-й строки = 2, 2-й строки = 1, 3-й строки = 2.

Число элементов в списке **width_ratios** и **height_ratios** должно соответствовать числу строк / столбцов. 

```python
fig = plt.figure(figsize=(10, 5))
gs = GridSpec(nrows=3, ncols=2, figure=fig, width_ratios = [1,2], height_ratios = [2,1,2])

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])
ax5 = fig.add_subplot(gs[2,0])
ax6 = fig.add_subplot(gs[2,1])

plt.tight_layout()
plt.show()
```

Оси в данной сетке могут быть объединены, например 1-я и 2-я строка. 

```python
fig = plt.figure(figsize=(10, 5))
gs = GridSpec(nrows=3, ncols=2, figure=fig, width_ratios = [1,2], height_ratios = [2,1,2])

ax1 = fig.add_subplot(gs[0:2,0])
ax2 = fig.add_subplot(gs[0:2,1])
ax3 = fig.add_subplot(gs[2,0])
ax4 = fig.add_subplot(gs[2,1])

plt.tight_layout()
plt.show()
```

Чтобы далее заполнить эти оси графиками, достаточно вызвать у этих осей методы, соответствующие графикам. 

```python
fig = plt.figure(figsize=(10, 5))
gs = GridSpec(nrows=3, ncols=2, figure=fig, width_ratios = [1,2], height_ratios = [2,1,2])

x = np.linspace(-np.pi, np.pi, 30)
y1 = np.sin(x)
y2 = 0.5 * np.sin(x)
y3 = 2 * np.sin(x)

ax1 = fig.add_subplot(gs[0:2,0])
ax1.plot(x,y1)

ax2 = fig.add_subplot(gs[0:2,1])
ax2.plot(y1,y2)

ax3 = fig.add_subplot(gs[2,0])
ax3.plot(y3,x)

ax4 = fig.add_subplot(gs[2,1])
ax4.plot(x,y3)

plt.tight_layout()
plt.show()
```

## Объект Scatter (диаграмма рассеяния)

Для построения диаграммы рассеяния в функцию **plt.scatter** необходимо передать два аргумента - значения по оси х и значения по оси y.

Дополнительные настройки:
* **s** - размер точки (определяется значение третьей переменной)
* **c** - цвет или цветовое представление точки (определяется исходя из значения вектора для данной точки и заданной цветовой гаммы **cmap**).
* **cmap** - цветовая гамма.
* **edgecolor** - цвет границы точки.
* **linewidth** - ширина границы точки.
* **alpha** - прозрачность точки. Очень полезна, когда необходимо определить зону перекрытия точек. По интенсивности цвета можно судить о наиболее сконцентрированной зоне.
* **marker** - отображение точки (маркер).

```python
from sklearn.datasets import load_iris

iris_dataset = load_iris()
df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset['feature_names'])
df['target'] = iris_dataset['target']
df['target'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

plt.figure(figsize = (16,3))
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'], c = '#ff7f0e')
plt.show()
```

Если необходимо отобразить диаграмму с точками разного цвета для разных групп, то создаётся несколько графиков - либо в цикле, либо как оси одного холста.

```python
groups = df.groupby('target')
x = 'petal length (cm)'
y = 'petal width (cm)'

plt.figure(figsize = (16,3))
for name, group in groups:
    plt.scatter(group[x], group[y], label=name)

plt.legend()
plt.show()
```

В аргумент **c** можно задать количественную переменную, диапазон которой вместе с указанием цветовой схемы **cmap** создадут градиент. Для однозначности определения лучше выводить цветовую легенду **colorbar**, если используется данный подход.

```python
colorgrad = df['petal width (cm)'] / df['sepal length (cm)']
plt.figure(figsize = (16,3))
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'], c = colorgrad, cmap = 'vanimo', marker = 'd')

cbar = plt.colorbar()
cbar.set_label('Отношение ширины \n чашелистика к его длине')
plt.show()
```

## Объекты Bar и Barh (столбчатая и линейчатая диаграммы)

Для базового построения столбчатой диаграммы необходимо передать в функцию два аргумента - значения по оси X и Y. В случае построения линейчатой диаграммы сначала передаются значения Y, затем Х. Другие настройки:
* **width** - ширина столбца;
* **color** - цвет столбца;
* **alpha** - прозрачность столбца.

```python
df_values = [['2019', 7238, 7239],
             ['2020', 7497, 7840],
             ['2021', 8997, 8410],
             ['2022', 9472, 9107]]
df_columns = ['Год', 'Торговые площади Магнит, тыс. кв.м.',
              'Торговые площади X5Group, тыс. кв.м.']
df = pd.DataFrame(df_values, columns=df_columns)

fig, axes = plt.subplots(1,2, figsize = (16,3))
axes[0].bar(df['Год'], df['Торговые площади X5Group, тыс. кв.м.'], color = 'darkred', alpha = 0.5)
axes[1].barh(df['Год'], df['Торговые площади X5Group, тыс. кв.м.'], color = 'lightgreen')
plt.show()
```

Поэтому для таких случаев лучше воспользоваться возможностями других библиотек. 

## Объект hist (гистограмма)

Гистограмма строится на основе одного обязательного аргумента - вектора значений, распределение которых необходимо отразить. 
Необязательные аргументы: 
* **bins** - число столбцов гистограммы (по умолчанию - 10). Если вместо одного числа указать список, то значения в этом списке интерпретируются как границы для построения столбцов (исключая правую границу, если она не последняя);
* **range** - границы диапазона, на основе которого будет строиться гистограмма;
* **cumulative** - возможные значения True / False / -1. Если не равно False, то отображается кумулятивная гистограмма.
* **orientation** - ориентация графика относительно оси Х (по умолчанию) или Y (orientation='horizontal').
* **log** - возможные значения True / False. Если True, шкала гистограммы пересчитывается в логарифмическую.
* **label** - подпись для легенды.

Другие настройки схожи с настройкой других графиков. 

```python
np.random.seed()
arr = np.random.randn(100000)

fig, axes = plt.subplots(1, 2, figsize = (16,3))
axes[0].hist(x=arr, color='green', cumulative=False)
axes[1].hist(x=arr, color='red', cumulative=True)
plt.show()
```

```python
np.random.seed()
arr = np.random.randn(100000)

fig, axes = plt.subplots(1, 2, figsize = (16,3))
axes[0].hist(x=arr, bins=20, color='lightgreen', cumulative=False, log=True, orientation='horizontal')
axes[1].hist(x=arr, bins=20, color='darkred', alpha=0.3, cumulative=True, log=True, orientation='horizontal')
plt.show()
```

## Объект boxplot (коробчатая диаграмма)

Обязательный аргумент функции один - вектор со значениями переменной. Необязательные аргументы:
* **patch_artist** - возможные значения: True / False. По умолчанию False. Добавляет оформление графика.
* **tick_labels** - список значений для подписи названия оси / группы / переменной, для которой строится график распределения.
* **orientation** - возможные значения: 'vertical' / 'horizontal'. По умолчанию 'vertical'. Регулирует отображение диаграммы с ориентацией по вертикали.
* **notch** - возможные значения: True / False. По умолчанию False. Регулирует отображение диаграммы с выемками, которые формируются в зависимости от доверительного интервала. 
* **whis** - значение для определения длины хвоста (количества IQR от 1-го и 3-го квартилей). По умолчанию - 1,5.
* **showmeans** - возможные значения: True / False. Регулирует отображение среднего значения на графике.

Для отрисовки нескольких графиков на одном холсте можно передать их как несколько столбцов одной таблицы.

```python
from sklearn.datasets import load_iris

iris_dataset = load_iris()
df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset['feature_names'])
df['target'] = iris_dataset['target']
df['target'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,3))
ax1.boxplot(df[df.columns[2]], tick_labels=[df.columns[2]], patch_artist=True, notch=True, orientation='horizontal', showmeans=True)
ax2 = plt.boxplot(df[df.columns[2:4]], tick_labels=df.columns[2:4], patch_artist=True, orientation='horizontal', whis = 0.3)
ax2['boxes'][0].set_facecolor('olive')
plt.tight_layout()
plt.show()
```

Объект **plt.boxplot** можно сохранять в отдельную переменную и в дальнейшем перезаписывать его объекты, исходя из того, что возвращает функция. Подробнее в [документации](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html). Например, в примере выше изменён цвет ящика на втором графике.

## Объект pie (круговая диаграмма)

Для построения круговой диаграммы достаточно передать один аргумент - столбец абсолютных значений по группам. Круговая диаграмма строится без дополнительных вычислений, визуализируя только то, что передано пользователем. Необязательные аргументы:
* **labels** - подписи секторов;
* **wedgeprops** - словарь со свойствами диаграммы (edgecolor - цвет границы сектора, linewidth - ширина границы, width - доля диаграммы, которая остаётся для отображения данных при преобраовании типа "бублик".
* **explode** - выделение доли отдельных секторов. В этот аргумент передаётся список значений с величиной сдвига каждого сектора относительно центра. Для тех секторов, которые не нужно сдвигать, значение указывается 0.
* **startangle** - угол поворота диаграммы.
* **colors** - принимает список цветов, в которые необходимо раскрасить сектора. Цвета могут быть переданы как названием, так и HEX-кодом.
* **radius** - размер диаграммы. По умолчанию равен 1. Если необходимо увеличить или уменьшить, то соответственно пишется значение относительно 1.
* **autopct** - в аргумент передаётся формат значений долей, которые будут отображаться внутри секторов.
* **pctdistance** - расположение подписи доли внутри секторов относительно внешней границы. По умолчанию значение равно 0,6. Если необходимо сдвигать ближе к центру, значение уменьшается, если дальше - увеличивается.
* **labeldistance** - расположение подписи сектора относительно его границы. По умолчанию значение равно 1,1. Если необходимо отдалить подписи, указывается более высокое значение.
* **counterclock** - возможные значения True / False. По умолчанию True - порядок следования секторов задаётся против часовой стрелки.

```python
df_values = [['Цех №1', 300],
             ['Цех №2', 210],
             ['Цех №3', 157],
             ['Цех №4', 180],
             ['Цех №5', 98]]
df_columns = ['Цех', 'Произведено деталей, шт']
df = pd.DataFrame(df_values, columns=df_columns)
```

```python
fig, axes = plt.subplots(1,3, figsize=(16,3))

axes[0].pie(df['Произведено деталей, шт'], labels=df['Цех'])
axes[1].pie(df['Произведено деталей, шт'], labels=df['Цех'], wedgeprops={'edgecolor': 'w', 'linewidth': 5, 'width': 0.6})
axes[2].pie(df['Произведено деталей, шт'], labels=df['Цех'], wedgeprops={'edgecolor': 'black', 'linewidth': 5, 'width': 0.3})
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize=(16,3))

axes[0].pie(df['Произведено деталей, шт'], labels=df['Цех'], explode = [0.2,0,0.1,0,0], labeldistance = 1.5)
axes[1].pie(df['Произведено деталей, шт'], labels=df['Цех'], startangle=45, autopct = '%1.1f%%', pctdistance = 0.7)
axes[2].pie(df['Произведено деталей, шт'], labels=df['Цех'], colors = ['blue', 'green', '#B7DC82', '#A4F6CB', '#C5CCAD'], counterclock = False)
plt.show()
```

## Объект stackplot (диаграмма с накоплением)

Для построения графика первым аргументом передаются значения по оси Х, а далее - значения по оси Y в том порядке, в котором их необходимо отобразить. Другие настройки:
* **baseline** - "исходная точка", относительно которой выравнивается график. По умолчанию - 'zero', т.е. обычная диаграмма с накоплением. Если задать значение 'sym', то график будет перестроен с центрированием относительно нуля.
* **colors** - список цветов для оформления каждой группы.
* **labels** - список подписей для легенды графика. 

```python
df_values = [['Понедельник', 300, 210, 157],
             ['Вторник', 295, 230, 180],
             ['Среда', 303, 214, 168],
             ['Четверг', 283, 254, 204],
             ['Пятница', 260, 263, 225],
             ['Суббота', 190, 177, 162],
             ['Воскресенье', 184, 181, 173]]
df_columns = ['День недели', 'Цех №1', 'Цех №2', 'Цех №3']
df = pd.DataFrame(df_values, columns=df_columns)
```

```python
fig, axes = plt.subplots(2,1, figsize=(16, 5))
axes[0].stackplot(df[df.columns[0]], df[df.columns[1]], df[df.columns[2]], df[df.columns[3]], labels=[1,2,3])
axes[1].stackplot(df[df.columns[0]], df[df.columns[1]], df[df.columns[2]], df[df.columns[3]], baseline='sym')
axes[0].legend()
plt.tight_layout()
plt.show()
```

## Сохранение графиков в файл

Для сохранения в файл используется команда **plt.savefig()**. 
Команда должна применяться до отображения графика **plt.show()**, когда в памяти ещё сохранён этот объект. После того, как график показывается, он закрывается, и сохраняться будет пустой холст.

В теле функции указывается название файла с его расширением (можно сохранять с разными расширениями - jpg / png / pdf / ...).
Название может быть расширенным, с указанием полного пути, куда должен быть сохранён график. 

Необязательным аргументом можно указать разрешение рисунка (аргумент **dpi**). Чтобы убрать лишние пустые границы, может быть использован аргумент **bbox_inches='tight'**. Есть много других необязательных настроек, часть которых может применяться только для формата jpg.

[Подробнее тут](https://matplotlib.org/3.1.3/api/_as_gen/matplotlib.pyplot.savefig.html).
